```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
prior_alpha = 1
prior_beta = 1
beta_dist = stats.beta(a = prior_alpha, b = prior_beta)

plt.figure(figsize=(10, 4))
plt.xlabel('Theta')
plt.ylabel('PDF')
plt.title('Prior Beta Distribtuion given prior_alpha={}, prior_beta={}'.format(prior_alpha,prior_beta))
x = np.linspace(0,1,500)
plt.plot(x, beta_dist.pdf(x))
```

```python
N = 141 
X = 107 

# posterior
posterior_alpha = prior_alpha + X
posterior_beta = prior_beta + N - X
post_beta_dist = stats.beta(a=posterior_alpha, b=posterior_beta)

#plotting 
x = np.linspace(0,1,500)
plt.figure(figsize=(10, 4))
plt.xlabel('Theta')
plt.ylabel('PDF')
plt.title('Posterior Beta Distribtuion given posterior_alpha={}, posterior_beta={}'.format(posterior_alpha,posterior_beta))
plt.plot(x, post_beta_dist.pdf(x))
```

The beta-binomial distribution has the following parameters.

* $n \in \mathbb{N}_0$ – the number of trials
* $\alpha, \beta \in \mathbb{R}^+$ – corresponding to the parameters of the beta distribution

The probability mass function of the beta-binomial distribution is

$$p(k | n,\alpha,\beta) = \binom{n}{k}\frac{B(k+\alpha,n-k+\beta)}{B(\alpha,\beta)}$$

where $B$ is the beta function. See the code below for how to evaluate the beta-binomial pmf in Python.
